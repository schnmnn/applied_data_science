```python
import pandas as pd
import re
import numpy as np
import scipy.stats as stats
```

```python
nhl_df=pd.read_csv("assets/nhl.csv")
nhl_df['year'] == 2018
cities=pd.read_html("assets/wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]
```

```python
nhl_df = nhl_df[nhl_df['year'] == 2018]
nhl_df['team'] = nhl_df['team'].str.replace('*', '', regex=False)
```

```python
def city_cleaning():
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    cities.rename(columns={"Population (2016 est.)[8]":'Population'},inplace=True)
    cities['NHL'] = cities['NHL'].str.replace(r'\[.*?\]', '', regex=True)
    cities = cities[(cities['NHL'] != '—') & (cities['NHL'] != '') ] 
    cities.loc[cities['NHL'] == 'Kings Ducks','NHL'] = 'Kings'
    cities.reset_index(inplace=True)
    
    return cities
```

```python
cities = city_cleaning()
```

```python
cities.loc[cities['NHL'] == 'Rangers Islanders Devils','NHL'] = 'Kings'
```

```python
nhl_df['team_list'] = nhl_df['team'].str.split()
```

```python
cities['NHL_list'] = cities['NHL'].str.split()
```

```python
def find_common_elements(list1, list2):
    common_elements = list(set(list1) & set(list2))
    return common_elements

result = []

for idx1, row1 in cities.iterrows():
    for idx2, row2 in nhl_df.iterrows():
        common_elements = find_common_elements(row1['NHL_list'], row2['team_list'])
        if common_elements:
            result.append({
                'cities_index': idx1, 
                'nhl_df_index': idx2, 
                'cities_values': row1['NHL_list'], 
                'nhl_df_values': row2['team_list'], 
                'common_elements': common_elements
            })

result_df = pd.DataFrame(result)

print("Result DataFrame:")
print(result_df)
```
